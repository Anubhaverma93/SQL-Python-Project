```python
import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#Changed str into lower & replaced space into underscore
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
#Create New Column to See discount
df['discount']=df['list_price']*df['discount_percent']/100
```

```python
#Create New Column to See sales price
df['sale_price']= df['list_price']-df['discount']
```

```python
#Create New Column to See profit
df['profit'] = df['sale_price']-df['cost_price']
```

```python
# convert order date from object to date datatype -->
# df.datatype
#pd.to_datetime(df['order_date'],format='%Y-%m-%d')
#changed datatype in exixting date column
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```
